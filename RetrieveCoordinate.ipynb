```python
import pandas as pd 
from datetime import datetime

#reading from file
receivedFileName="resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv"
resaleFlatDF = pd.read_csv(receivedFileName, encoding = "ISO-8859-1")
resaleFlatDF
```

```python
import requests

resaleFlatDF["Latitude"] = ""
resaleFlatDF["Longitude"] = ""
startPoint = 0
for i in range(startPoint, startPoint+10000):
    lat = ""
    lng = ""
    destination = resaleFlatDF.at[i,"block"] +" "+ resaleFlatDF.at[i,"street_name"] 
    urlAPI = 'https://developers.onemap.sg/commonapi/search?searchVal='+destination+'&returnGeom=Y&getAddrDetails=Y'
    r = requests.get(urlAPI)
    result = r.json()
    if result["results"] != []:
        lat = result["results"][0]["LATITUDE"]
        lng = result["results"][0]["LONGITUDE"]
    resaleFlatDF.at[i,"Latitude"] = lat
    resaleFlatDF.at[i,"Longitude"] = lng
```

```python
resaleFlatDF.to_csv("DataAnalytics.csv", index=False)
```
